### Functions definitions

In this notebook we reuse the same helper functions already introduced in the previous one, such as load_model for restoring trained models and the various create_synthetic_data_* functions for generating synthetic samples with each approach.

```python
def load_model(file_path):
  with(open(file_path, 'rb')) as f:
    model = pickle.load(f)
  return model
```

```python
def create_synthetic_data_CopulaGAN(model, num_rows):
    synthetic_data = model.sample(num_rows=num_rows)
    return synthetic_data
```

```python
def create_synthetic_data_TVAE(model, num_rows):
    synthetic_data = model.sample(num_rows=num_rows)
    return synthetic_data
```

```python
def create_synthetic_data_CTGAN(model, num_rows):
    synthetic_data = model.sample(num_rows=num_rows)
    return synthetic_data
```

```python
def create_synthetic_data_distil_GReaT(model, num_rows):
    synthetic_data = model.sample(n_samples=num_rows, max_length=2000)
    return synthetic_data
```

```python
def create_synthetic_data_GReaT(model, num_rows):
    synthetic_data = model.sample(n_samples=num_rows, max_length=2000)
    return synthetic_data
```

## 1 Sampling

In this case, we no longer observe the presence of "out-of-context words" that appeared in the synthetic data generated with GReaT for the German Credit Risk dataset. This difference arises because all the columns in the Australian Credit Approval dataset are numerical: even the categorical features were originally encoded as numbers to preserve privacy. Unlike in the German dataset, where LLM-based models such as GReaT tended to generate extra text fragments due to their natural predisposition to produce free-form sequences, the numerical representation here prevents such anomalies. While this characteristic helps generative models produce cleaner and more consistent outputs, it also makes it harder for a human to directly assess the quality and coherence of the synthetic data, shifting the focus toward quantitative evaluation metrics.
